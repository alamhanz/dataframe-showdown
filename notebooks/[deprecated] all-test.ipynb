```python
import os
os.environ["MODIN_ENGINE"] = "dask"
import pandas as pd
import polars as pl 
import dask.dataframe as dd
import vaex as vx
import modin.pandas as md
from dask.distributed import Client

import time
import numpy as np

import shutil
```

```python
PATH_DATA = '../../df-showdown-data/'
TEMP_PATH = PATH_DATA+'temp/'
PATH_OUTPUT = '../output/'
```

```python
# https://modin.readthedocs.io/en/stable/getting_started/why_modin/out_of_core.html
cli = Client()
cli
```

## Input

```python
all_sla = {}
all_sla['pandas'] = []
all_sla['modin_dask'] = []
all_sla['polars'] = []
all_sla['dask'] = []
all_sla['vaex'] = []
all_sla['rows_cnt_in_million'] = []

for i in range(8):
    fn = 'data_{}.csv'.format(i)
    
    start = time.perf_counter()
    dtemp1 = pd.read_csv(PATH_DATA+fn)
    end = time.perf_counter() - start
    all_sla['pandas'].append(end)

    nrow = round(dtemp1.shape[0]/1000000,2)
    all_sla['rows_cnt_in_million'].append(nrow)
    del dtemp1

    start = time.perf_counter()
    dtemp1 = md.read_csv(PATH_DATA+fn)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['modin_dask'].append(end)

    start = time.perf_counter()
    dtemp1 = pl.read_csv(PATH_DATA+fn)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['polars'].append(end)

    start = time.perf_counter()
    dtemp1 = dd.read_csv(PATH_DATA+fn)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['dask'].append(end)

    start = time.perf_counter()
    dtemp1 = vx.open(PATH_DATA+fn)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['vaex'].append(end)

    print(i)

all_sla_csv = all_sla.copy()
```

```python
df_sla1 = pd.DataFrame(all_sla_csv)
df_sla1.to_csv(PATH_OUTPUT+'sla_input-csv.csv',index = False)
df_sla1
```

## Output

```python
all_sla = {}
all_sla['pandas'] = []
all_sla['polars'] = []
all_sla['dask_default'] = []
all_sla['dask_single'] = []
all_sla['vaex'] = []
all_sla['rows_cnt_in_million'] = []

for i in range(9):
    fn = 'data_{}.csv'.format(i)

    dtemp1 = pd.read_csv(PATH_DATA+fn)
    start = time.perf_counter()
    dtemp1.to_csv(TEMP_PATH + 'temp.csv',index=False)
    end = time.perf_counter() - start
    all_sla['pandas'].append(end)

    nrow = round(dtemp1.shape[0]/1000000,2)
    all_sla['rows_cnt_in_million'].append(nrow)
    del dtemp1
    os.remove(TEMP_PATH + 'temp.csv')

    dtemp1 = pl.read_csv(PATH_DATA+fn)
    start = time.perf_counter()
    dtemp1.write_csv(TEMP_PATH + 'temp.csv')
    end = time.perf_counter() - start
    del dtemp1
    all_sla['polars'].append(end)
    os.remove(TEMP_PATH + 'temp.csv')

    # https://mungingdata.com/dask/ouput-dataframe-single-csv-file/
    dtemp1 = dd.read_csv(PATH_DATA+fn)
    start = time.perf_counter()
    dtemp1.to_csv(TEMP_PATH + 'temp_dask/',index=False)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['dask_default'].append(end)
    shutil.rmtree(TEMP_PATH + 'temp_dask/')

    dtemp1 = dd.read_csv(PATH_DATA+fn)
    start = time.perf_counter()
    dtemp1.compute().to_csv(TEMP_PATH + 'temp.csv',index=False)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['dask_single'].append(end)
    os.remove(TEMP_PATH + 'temp.csv')

    dtemp1 = vx.open(PATH_DATA+fn)
    start = time.perf_counter()
    dtemp1.export(TEMP_PATH + 'temp.csv',index=False)
    end = time.perf_counter() - start
    del dtemp1
    all_sla['vaex'].append(end)
    os.remove(TEMP_PATH + 'temp.csv')

    print(i)

all_sla_csv = all_sla.copy()
```

```python
df_sla1 = pd.DataFrame(all_sla_csv)
df_sla1.to_csv(PATH_OUTPUT+'sla_output-csv.csv',index = False)
df_sla1
```
